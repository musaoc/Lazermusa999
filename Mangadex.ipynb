```python
import os
import re


# Define the folder path in Google Drive
folder_path = '/content/Goodnight Punpun'  # Replace with your folder path

# Create a list of all files in the folder
files = os.listdir(folder_path)

# Regular expression to extract volume number
volume_pattern = re.compile(r'Vol\. (\d+)')

# Create folders and organize files
for file in files:
    # Skip directories (if any)
    if os.path.isdir(os.path.join(folder_path, file)):
        continue

    # Search for volume number in the filename
    match = volume_pattern.search(file)

    if match:
        # Extract volume number
        volume_number = match.group(1)
        volume_folder = f'Vol. {volume_number}'

        # Create volume folder if it doesn't exist
        volume_folder_path = os.path.join(folder_path, volume_folder)
        if not os.path.exists(volume_folder_path):
            os.makedirs(volume_folder_path)

        # Move file to the volume folder
        src = os.path.join(folder_path, file)
        dest = os.path.join(volume_folder_path, file)
        os.rename(src, dest)
    else:
        # Create 'Other' folder if it doesn't exist
        other_folder_path = os.path.join(folder_path, 'Other')
        if not os.path.exists(other_folder_path):
            os.makedirs(other_folder_path)

        # Move file to the 'Other' folder
        src = os.path.join(folder_path, file)
        dest = os.path.join(other_folder_path, file)
        os.rename(src, dest)

print("Files organized successfully!")
```
